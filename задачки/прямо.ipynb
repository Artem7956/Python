```python
import numpy as np
#N = 4
#M = 4

A=  np.matrix([
    [0,0,0],
    [1,0,0],
    [1,1,0]
    ])

#A=A*0
N,M=A.shape[0],A.shape[1]

#import numpy as np
def is_rectangle(cells):
    cells.sort()
    i=0
    side=1
    
    if len(cells)==0:
        return False
    elif len(cells)==1:
        return True
        
    while i< len(cells)-1:
        if cells[i+1]-cells[i]==1 and cells[i]%M>0:
            side=side+1
        else:
            break
        i=i+1    
    if  len(cells)%side > 0:
        return False
       
    #s=sum(cells[0*side:(0)*side+side])
    #print('side',side,s,sum(cells[1*side:(1)*side+side]),1*side*M)
           
    for k in range(0,len(cells)//side-1):
        
        difs=np.array(cells[(k+1)*side:(k+1)*side+side])-np.array(cells[(k)*side:(k)*side+side])
        #print(k,side,difs)
        if not (difs==np.ones(side)*M).all():
            return False
        #print(np.ones(2)==np.ones(2)*2)    
          
        #if sum(cells[k*side:(k)*side+side])-side*M != s:
           
           #return False
           
    return True          

#print(is_rectangle([2,3,4,7]))    
def subsets( S ):
    sets = []
    len_S = len( S )
    for i in range( 1 << len_S ):
        print(i)
        subset = [ S[ bit ] for bit in range( len_S ) if i & ( 1 << bit ) ]
        if is_rectangle(subset):
            sets.append( subset)
    return sets

```
